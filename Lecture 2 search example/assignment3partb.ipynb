```python
from inverted_index import InvertedIndex
from utils import read_data
inv_ind = InvertedIndex()
documents = read_data('./shakespeare')
for d in documents:
    print(d[0])
    inv_ind.add_document(d)
```

```python
inv_ind.calcLogEntropy()
inv_ind.generate_term_by_doc_matrix(log_entropy=True)
```

```python
results_euclidean = inv_ind.search_euclidean("scotland kings and thanes")
# sort and print only top 10 results
print("Top 10 Results using Euclidean distance:\n")
results_euclidean.sort(key=lambda x: x[1], reverse=True)
for i in range(10):
    print(results_euclidean[i])

results_euclidean_idf = inv_ind.search_euclidean("scotland kings and thanes", tfidf=True)
print("\nTop 10 Results using Euclidean distance with tfidf:\n")
results_euclidean_idf.sort(key=lambda x: x[1], reverse=True)
for i in range(10):
    print(results_euclidean_idf[i])

results_euclidean_entropy = inv_ind.search_euclidean("scotland kings and thanes", log_entropy=True)
print("\nTop 10 Results using Euclidean distance with log entropy:\n")
results_euclidean_entropy.sort(key=lambda x: x[1], reverse=True)
for i in range(10):
    print(results_euclidean_entropy[i])
```
